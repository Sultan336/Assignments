```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob 
```

```python
# using the glob libraary to read all csv files
census_data = glob.glob('states[0-9].csv') # states*.csv and states?csv also work
census_data                                #now we have a list of all the files
```

```python
combined_data = []                      #storing all file data in one variable
for all_files in census_data:
    raw_data = pd.read_csv(all_files)
    combined_data.append(raw_data)
combined_data

```

```python
us_census = pd.concat(combined_data,  ignore_index=True) 
us_census.head()
```

```python
us_census.info()
```

```python
us_census.describe()
```

```python
us_census.columns
```

```python
us_census.drop('Unnamed: 0', inplace=True, axis=1) #reomving the Unnamed column
```

```python
us_census.head()
```

```python
us_census['Income'] = us_census['Income'].replace(['\$'],'', regex=True) # Reomving the dollar sign
us_census.Income.head()
```

```python
us_census['Income'] = pd.to_numeric(us_census.Income)  #converting string into a number
```

```python
us_census.Income.head()
```

```python
us_census['Income'].head()
```

```python
us_census['Men'] = us_census['GenderPop'].str.split('(_)',expand=True)[0]   #spliting genderprop into two columns
us_census['Women'] = us_census['GenderPop'].str.split('(_)',expand=True)[2]
us_census.head()
```

```python
us_census['Men'] = us_census['Men'].replace(['M'],'', regex=True)       #removing M anf F and converting into numeric data 
us_census['Women'] = us_census['Women'].replace(['F'],'', regex=True)
us_census['Men'] = pd.to_numeric(us_census.Men)
us_census['Women'] = pd.to_numeric(us_census.Women)
us_census.head()
```

```python
#Plotting Men's and Women's Income
plt.scatter(us_census.Women,us_census.Income)
plt.xlabel('Women')
plt.ylabel('Income')
plt.show()
```

```python
#Create functionm to plot different columns
def print_histogram(df, columns):
  for column in columns:
    plt.hist(df[column])
    plt.title('Figure : ' + column)
    plt.show()
    plt.clf()

# Create a list of columns to plot
columns = ['Hispanic', 'White', 'Black', 'Native', 'Asian', 'Pacific']
```

```python
# Task 3  The first 10 rows represent data from your Staten Island location. 
#Select these rows and save them to staten_island.
staten_island = inventory.loc[:9]
staten_island
```

```python
#Task 4   A customer just emailed you asking what products are sold at your Staten Island location.
#Select the column product_description from staten_island 
#and save it to the variable product_request.

product_request = staten_island.product_description
product_request
```

```python
# Task 5 Another customer emails to ask what types of seeds are sold at the Brooklyn location.
seed_request = inventory[(inventory.location == 'Brooklyn') & (inventory.product_type == 'seeds')]
seed_request
```

```python
# Task 6   Add a column to inventory called in_stock which is True 
# if quantity is greater than 0 and False if quantity equals 0.
# inventory['in_stock'] = np.where(inventory.quantity == 0, False,True) or
inventory['in_stock'] = inventory.quantity > 0
inventory
```

```python
# Task 7  Petal Power wants to know how valuable their current inventory is
inventory['total_value'] = inventory.price * inventory.quantity
inventory
```

```python
# Task 8   The Marketing department wants a complete description of each product for their catalog.

#The following lambda function combines product_type and product_description into a single string:
combine_lambda = lambda row: \
    '{} - {}'.format(row.product_type,
                     row.product_description)
# Task 9
inventory['full_description'] = inventory.apply(combine_lambda,axis=1)
inventory
```
