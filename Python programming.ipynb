```python
url = requests.get('https://keithgalli.github.io/web-scraping/webpage.html')
webpage = BeautifulSoup(url.content)
print(webpage.prettify())
```

```python
# grabing all links in 3 different ways
```

# 1 select

```python
links = webpage.select('ul.socials a')
actual_links = [link['href'] for link in links]
actual_links
```

# 2 find & fina_all

```python
ulist = webpage.find('ul',attrs={'class':'socials'})
links = ulist.find_all('a')
actual_links = [link ['href'] for link in links]
actual_links
```

# 3

```python
links = webpage.select('li.social a')
actual_links = [link['href'] for link in links]
actual_links
```

```python
#Scraping the table
import pandas as pd
```

```python
table = webpage.select('table.hockey-stats')[0]
columns = table.find('thead').find_all('th')
column_names = [c.string for c in columns]
table_rows = table.find('tbody').find_all('tr')
rows = []
for tr in table_rows:
    td = tr.find_all('td')
    row = [str(tr.get_text()).strip() for tr in td]
    rows.append(row)
print(rows[0])
df = pd.DataFrame(rows,columns=column_names)
df['S']
```

```python
# Grabing all the fun facts that contain "is"
```

```python
fun_facts = webpage.select('ul.fun-facts li')
facts_with_is = [fact.find(string=re.compile('is'))for fact in fun_facts]
facts_with_is = [fact.find_parent().get_text() for fact in facts_with_is if fact]
facts_with_is
```
